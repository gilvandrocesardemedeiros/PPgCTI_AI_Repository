```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataSet = pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
```

```python
dataSet.corr()
```

```python
dataSet.head()
```

```python
X = dataSet.drop(columns = "variety")
y = dataSet["variety"]
```

```python
y.drop_duplicates()
```

```python
y.loc[y == "Setosa"] = 0
y.loc[y == "Versicolor"] = 1
y.loc[y == "Virginica"] = 2
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X["sepal.length"], X["petal.width"], c=y)
plt.show()
```

```python
from lpproj import LocalityPreservingProjection
lpp = LocalityPreservingProjection(n_components=2)
X_2D = lpp.fit_transform(X)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X_2D[:, 0], X_2D[:, 1], c=y)
plt.title("Redução de dimensionalidade usando LPP")
plt.show()
```

```python
from sklearn.decomposition import PCA
plt.figure(figsize=(8,8))
Xpca = PCA(n_components=2).fit_transform(X)
plt.title("Redução de dimensionalidade usando PCA")
plt.scatter(Xpca[:, 0], Xpca[:, 1], c=y)
```
